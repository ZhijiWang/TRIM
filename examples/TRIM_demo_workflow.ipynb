# TRIM Demo Workflow

This notebook runs the core TRIM demonstration workflow: loading annotations, validating schema conformance, comparing signatures, and building a corpus graph.

```python
from pathlib import Path

import pandas as pd

from trim.compare import same_function_table, same_cue_table, broad_family_table, contested_cases_table
from trim.graph import build_corpus_graph, graph_summary
from trim.validator import validation_report

repo_root = Path.cwd()
data_path = repo_root / "data" / "demo_annotations.csv"
```

## Load Demonstration Annotations

```python
df = pd.read_csv(data_path, dtype=str, keep_default_na=False)
df.head()
```

## Basic Corpus Counts

```python
summary_counts = {
    "number_of_cases": len(df),
    "counts_by_source": df["source"].value_counts().to_dict(),
    "counts_by_friction_locus": df["friction_locus"].value_counts().to_dict(),
    "counts_by_rationale_mechanism": df["rationale_mechanism"].value_counts().to_dict(),
}
summary_counts
```

## Validate Annotations

```python
issues = validation_report(df)
issues
```

## Comparison Tables

```python
same_function_table(df)
```

```python
same_cue_table(df)
```

```python
broad_family_table(df)
```

```python
contested_cases_table(df)
```

## Graph Summary

```python
G = build_corpus_graph(df)
graph_summary(G)
```
